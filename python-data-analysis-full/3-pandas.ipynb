# 3. Pandas

**Pandas is powerful and easy-to-use library for data analysis. Is has two main object to represents data: Series and DataFrame.**

```python
# import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = pd.Series([1,2,3,4,5])
x
```

```python
# Basic Operation on the Series

x + 100
```

```python
(x **2) + 100
```

```python
x > 2
```

# apply()

```python
def f(x):
    if x % 2 == 0:
        return x*2
    else:
        return x*3
    
x.apply(f)
```

# astype()

```python
x.astype(np.float64)
```

# DataFrame

```python
data = [1,2,3,4,5,6,7]
df = pd.DataFrame(data, columns=["x"])
```

```python
df
```

```python
df["x"]
```

```python
df["x"][0]
```

```python
# Add the extra columns

df["x_plus_2"] = df["x"] + 2
```

```python
df
```

# map()

```python
df["is_even"] = df["x"]%2
```

```python
df
```

```python
df["odd_even"] = df["is_even"].map({1:"odd", 0:"even"})
```

```python
df
```

# drop()

```python
df = df.drop("is_even", 1)
```

```python
df
```

```python
df["x_square"] = df["x"]**2
```

```python
df
```

```python
df["x_factorial"] = df["x"].apply(np.math.factorial)
```

```python
df
```

# Filtering

```python
df[df["odd_even"] == "odd"]
```

```python
df[df["odd_even"] == "even"]
```

# ScatterPlot

```python
pd.scatter_matrix(df, figsize=(10,10));
```
