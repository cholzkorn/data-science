# Running a Random Forest


### Importing the required libraries

```python
%matplotlib inline

from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import os
import matplotlib.pylab as plt
from sklearn.cross_validation import train_test_split
import sklearn.metrics
import seaborn as sns
```

```python
os.chdir('C:/Users/shivam/Desktop')
iris = pd.read_csv("iris.csv")
```

```python
iris.head()
```

## Exploratory Data Analysis

```python
sns.pairplot(data = iris)
```

### Split the datset into training and testing dataset


```python
feature_cols = ["Petal.Length","Sepal.Length", "Petal.Width","Sepal.Width"]
X = iris[feature_cols]
y = iris.Species

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =.25)
```

## Build Model on training dataset 

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators = 25)
classifier.fit(X_train,y_train)
```

```python
predictions = classifier.predict(X_test)
```

```python
conf_matrix = sklearn.metrics.confusion_matrix(y_test,predictions)
```

```python
conf_matrix
```

```python
sklearn.metrics.accuracy_score(y_test,predictions)
```
