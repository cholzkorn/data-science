# Importing packages and data

```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
# Renaming columns
df = df.rename(index=str, columns={"Rank": "rank",
                                   "Grade": "grade",
                                   "Channel name": "channel",
                                   "Video Uploads": "uploads",
                                   "Subscribers": "subscribers",
                                   "Video views": "views"})
df.head()
```

```python
# checking for NAs
df.isna().sum()
```

```python
# checking types
df.dtypes
```

```python
# replacing omitted values
df.uploads = df.uploads.replace("None", None)
df.subscribers = df.subscribers.replace("None", None)
df.views = df.views.replace("None", None)

# Transforming types
df.uploads = pd.to_numeric(df.uploads)
df.subscribers = pd.to_numeric(df.subscribers)
df.views = pd.to_numeric(df.views)

df.grade = df["grade"].astype("category")
```

```python
# checking types again
df.dtypes
```

# Feature engineering

```python
# Creating new features

df["views_per_video"] = df["views"] / df["uploads"]
df["views_per_subscriber"] = df["views"] / df["subscribers"]

# words = number of words in channel name, special signs also counted as words

df["words"] = df["channel"].str.lower().str.split().apply(lambda x: len(x))
df.head()
```

```python
# Is Gaming in the channel title?

df["gaming"] = np.where(df["channel"].str.contains("game*|gaming",
                                                    case=False,
                                                    na=False), True, False)
df
```

# Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Correlation matrix
corr = df.corr()
hm = sns.heatmap(corr,
           xticklabels=corr.columns.values,
           yticklabels=corr.columns.values,
           annot=True)
hm.set_title("Correlation Matrix")
```

```python
# Which grade has the most views per video?
vpv_box = sns.boxplot(x=df["grade"], y=pd.to_numeric(df.views_per_video),
                      palette="Set3")
```

```python
# Since that doesn't give us too much information, let's only look at the
# top 100

df30 = df.iloc[0:30]

vpv_box30 = sns.boxplot(x=df30["grade"], y=pd.to_numeric(df30.views_per_video),
                        palette="Set3")
```

```python
# A+ channels seem to have more views per video than A++ videos.
# Maybe they just have more videos on average?

df_uploads_mn = df.groupby(["grade"], as_index=False).mean()
df_uploads_mn.sort_values("uploads", ascending=False)
```

```python
# Which channels have the most views per subscriber?

df.sort_values(by="views_per_subscriber", ascending=True).head()
```

```python
# How many of the top 5000 channels have game* or gaming in their title?
((df[df["gaming"] == True]).count()).gaming
```

# Findings

```python
# CORRELATIONS

# Total subscribers and total views are, as expected, heavily correlated

# The correlation between subscribers and views
# per subscriber is negligible. This means that in our data, smaller and bigger
# channels could have similar viewer patterns.
```

```python
# GRADES - (everything on average)

# A++ and A- grade have got the most views per video, while A++ have also got
# the most uploads.

# A grade channels have the most subscribers and the most views per subscriber.
```
